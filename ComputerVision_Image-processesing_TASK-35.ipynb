# Successfully Created a Gif 

```python
# importing the modules
import numpy as np
import cv2

# creating array using np.zeroes()
array = np.zeros([500, 500, 3],dtype = np.uint8)

# setting RGB color values as 255,255,255
j=50
k=0
array[:, :] = [150, 255, 0]
for i in array:
        if i.any()<=255:
                                                         # displaying the image
            array[:,:] = [150,255-j,k]
            k=k+10
            j=j-50
            cv2.imshow("image", array)
            if cv2.waitKey(10)==13:
                break
        else:
            break

cv2.destroyAllWindows()
```

# Images piled side by side, horizantal & vertically

```python
import numpy as np
import cv2, time
i=0
l=[]
def img(x,y,z):
    array2 = np.zeros([50,50,3],dtype = np.uint8)  
    array2[:,:]=[x,y,z]
    #cv2.imwrite("Image1Created.jpg", array)
    #pic = cv2.cvtColor(array2.astype(np.uint8), cv2.COLOR_BGR2RGB)
    l.append(array2)
    
img(10,255,255)
img(150,20,300)
img(255,100,150)
img(255,255,200)
#print(l)

listimg = np.concatenate((l[0],l[1],l[2]),axis=1)
cv2.imshow('List of Images --->> Horizanally',listimg)

cv2.waitKey()
cv2.destroyAllWindows()



listimg1 = np.concatenate((l[0],l[1],l[2]),axis=0)
cv2.imshow('List of Images --->> Vertically',listimg1)

cv2.waitKey()
cv2.destroyAllWindows()
```

# Resizing the Images

```python
import numpy as np
import cv2, time

array = np.zeros([350,350,3],dtype = np.uint8)  
array[:,:]=[150,255,50]
cv2.imshow("img1",array)
cv2.waitKey()
cv2.destroyAllWindows()

# Resizing the image
array2 = cv2.resize(array, (255, 255))
cv2.imshow("img1",array2)
cv2.waitKey()
cv2.destroyAllWindows()
```

# Cropping of images

```python

import numpy as np
import cv2

arrayy = np.zeros([350,350,3],dtype = np.uint8)                      # With Datatype

arrayy[:,:]=[255,50,255]
cv2.imshow('Image', arrayy)
cv2.waitKey()
cv2.destroyAllWindows()


y=0
x=0
h=100
w=200
crop = array[y:y+h, x:x+w]
cv2.imshow('Image', crop)
cv2.waitKey()
cv2.destroyAllWindows()
```
